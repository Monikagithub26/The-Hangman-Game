# THE HANGMAN GAME

## The Hangman program randomly selects a secret word from a list of secret words. The random module will provide this ability, so line 1 in program imports it. 
## The Game: Here, a random word (a fruit name) is picked up from our collection and the player gets limited chances to win the game.
## When a letter in that word is guessed correctly, that letter position in the word is made visible. In this way, all letters of the word are to be guessed before all the chances are over.
## For convenience, we have given length of word + 2 chances. For example, word to be guessed is mango, then user gets 5 + 2 = 7 chances, as mango is a five letter word

```python
import random
from collections import Counter

print("....HANGMAN GAME....")
name = input("Enter your Name -> ")
print("Welcome",name,"!")
print("Try to guess the word before chances get over")
# Collection of words (word will be chosen at random from these)
someWords = '''apple banana mango strawberry orange grape pineapple apricot lemon coconut watermelon
cherry papaya berry peach lychee muskmelon avocado cantaloupe blackplum dates dragonfruit hackberry 
guava pomegranate sapota kiwi pear figs lime bear sloe custardapple olive mulberry blackberry blackcurrent
blueberry breadfruit gooseberry jackfruit peach pomelo quince malayapple rasppberry sapodilla alum ziziphus
surinamcherry tamarillo kiwano feijoa woodapple palmyrafruit'''
someWords = someWords.split(' ')
# randomly choose a secret word from our "someWords" LIST.
word = random.choice(someWords)
if __name__ == '__main__':
    print('Guess the word! HINT: word is a name of a fruit')
    for i in word:
        # For printing the empty spaces for letters of the word
        print('_', end=' ')
    print()
    playing = True
    # list for storing the letters guessed by the player
    letterGuessed = ''
    chances = len(word) + 2
    correct = 0
    flag = 0
    try:
        while (chances != 0) and flag == 0:  # flag is updated when the word is correctly guessed
            print()
            # chances -= 1
            try:
                guess = str(input('Enter a letter to guess: '))
            except ValueError :
                print('Enter only a letter!')
                continue
            # Validation of the guess
            if not guess.isalpha():
                print('Enter only a LETTER')
                continue
            elif len(guess) > 1:
                print('Enter only a SINGLE letter')
                continue
            elif guess in letterGuessed:
                print('You have already guessed that letter')
                continue
            # If letter is guessed correctly
            if guess in word:
                k = word.count(guess)  # k stores the number of times the guessed letter occurs in the word
                for _ in range(k):
                    letterGuessed += guess    # The guess letter is added as many times as it occurs 
            if guess not in word:
                if chances == 9:
                    chances = chances - 1
                    print("9 turns left!!!")
                    print("--------------")
                elif chances == 8:
                    chances = chances - 1
                    print("8 turns left!!!")
                    print("--------------")
                    print("       o      ")
                elif chances == 7:
                    chances = chances - 1
                    print("7 turns left!!!")
                    print("--------------")
                    print("       O      ")
                    print("       |      ")
                elif chances == 6:
                    chances = chances - 1
                    print("6 turns left!!!")
                    print("--------------")
                    print("       O      ")
                    print("       |      ")
                    print("      /      ")
                elif chances == 5:
                    chances = chances - 1
                    print("5 turns left!!!")
                    print("--------------")
                    print("       O      ")
                    print("       |      ")
                    print("      / \      ")
                elif chances == 4:
                    chances = chances - 1
                    print("4 turns left!!!")
                    print("--------------")
                    print("      \O      ")
                    print("       |      ")
                    print("      / \      ")
                elif chances == 3:
                    chances = chances - 1
                    print("3 turns left!!!")
                    print("--------------")
                    print("      \O/      ")
                    print("       |      ")
                    print("      / \      ")
                elif chances == 2:
                    chances = chances - 1
                    print("2 turns left!!!")
                    print("--------------")
                    print("      \O/ |     ")
                    print("       |      ")
                    print("      / \      ")
                elif chances == 1:
                    chances = chances - 1
                    print("only 1 turn left!!! hangman on his last breath")
                    print("--------------")
                    print("      \O/_|     ")
                    print("       |      ")
                    print("      / \      ")
                else:
                    print("You loose")
                    print("You let a good man die")
                    print("--------------")
                    print("       O_|     ")
                    print("      /|\      ")
                    print("      / \      ")
                    break
            # Print the word
            for char in word:
                if char in letterGuessed and (Counter(letterGuessed) != Counter(word)):
                    print(char, end=' ')
                    correct += 1
                # If user has guessed all the letters
                elif (Counter(letterGuessed) == Counter(word)):  # Once the correct word is guessed fully,
                    # the game ends, even if chances remain
                    print("The word is: ", end=' ')
                    print(word)
                    flag = 1
                    print('\nCongratulations, You won!')
                    break  # To break out of the for loop
                else:
                    print('_', end=' ')
        # If user has used all of his chances
        if chances <= 0 and (Counter(letterGuessed) != Counter(word)):
            print()
            print('You lost! Try again..')
            print("You let a good man die")
            print("--------------")
            print("       O_|     ")
            print("      /|\      ")
            print("      / \      ")
            print('The word was {}'.format(word))
            print("Game over!")
    except KeyboardInterrupt:
        print()
        print('Better! Try again.')
        exit()
```
